#Некорректная идентификация столбцов

Некоторые столбы имеют некорректные названия, по которым непонятно какие именно данные в них представлены. Среди них:

-Parch (столбец обозначающий наличие на борту родителей/детей) - записан как целое слово, а не аббревиатура, значение непонятно

-Embarked (столбец, обозначающий порт посадки) - больше бы подошло название "port"


Переименуем эти столбцы:


```python
df_titanic = df_titanic.rename(columns = {'Parch': 'Par_Ch'})
```

```python
df_titanic = df_titanic.rename(columns = {'Embarked': 'Port'})
```

#Графики

```python
import matplotlib.pyplot as plt
count_dict = {}
for s, p in zip(df_titanic['Survived'], df_titanic['Pclass']):
    if (s, p) in count_dict:
        count_dict[(s, p)] += 1
    else:
        count_dict[(s, p)] = 1

# Разделяем данные на выживших и погибших
survived_count = [count_dict.get((1, 'A'), 0), count_dict.get((1, 'B'), 0), count_dict.get((1, 'C'), 0)]
dead_count = [count_dict.get((0, 'A'), 0), count_dict.get((0, 'B'), 0), count_dict.get((0, 'C'), 0)]

# Создаем график
labels = ['A', 'B', 'C']
x = range(len(labels))

fig, ax = plt.subplots()
ax.bar(x, survived_count, label='Выжили')
ax.bar(x, dead_count, bottom=survived_count, label='Погибли')

# Задаем оформление графика
ax.set_xlabel('Класс')
ax.set_ylabel('Количество пассажиров')
ax.set_title('Распределение выживших и погибших пассажиров по классам')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Отображаем график
plt.show()
```
