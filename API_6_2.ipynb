```python
import matplotlib.pyplot as plt
count_dict = {}
for s, p in zip(df_titanic['Survived'], df_titanic['Pclass']):
    if (s, p) in count_dict:
        count_dict[(s, p)] += 1
    else:
        count_dict[(s, p)] = 1

#      
survived_count = [count_dict.get((1, 'A'), 0), count_dict.get((1, 'B'), 0), count_dict.get((1, 'C'), 0)]
dead_count = [count_dict.get((0, 'A'), 0), count_dict.get((0, 'B'), 0), count_dict.get((0, 'C'), 0)]

#  
labels = ['A', 'B', 'C']
x = range(len(labels))

fig, ax = plt.subplots()
ax.bar(x, survived_count, label='')
ax.bar(x, dead_count, bottom=survived_count, label='')

#   
ax.set_xlabel('')
ax.set_ylabel(' ')
ax.set_title('      ')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

#  
plt.show()
```
